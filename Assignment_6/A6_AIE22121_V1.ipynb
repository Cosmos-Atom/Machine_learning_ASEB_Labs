```python
# importations of libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder




def Question_1(input_ch, output, a):
    X = input_ch

    y = output

    W = [10, 0.2, -0.75]

    alpha = a

    def step_function(x):
        return 1 if x >= 0 else 0
    
    def predict(X, W):
        z = np.dot(X, W[1:]) + W[0]
        return step_function(z)

    
    def train_step(X, y, W, alpha, epochs):
        errors = []
        for epoch in range(epochs):
            sum_squared_error = 0
            for i in range(len(X)):
                prediction = predict(X[i], W)
                error = y[i] - prediction
                sum_squared_error += error ** 2
                W[1:] += alpha * error * X[i]
                W[0] += alpha * error
            errors.append(sum_squared_error)
            if sum_squared_error <= 0.002:
                break
        return W, errors
    
    epochs = 1000
    trained_weights, errors = train_step(X, y, W, alpha, epochs)
    plt.plot(range(1, len(errors) + 1), errors)
    plt.xlabel('Epochs')
    plt.ylabel('Sum-Square Error')
    plt.title('Epochs vs. Sum-Square Error (Step)')
    plt.show()
    print("Trained Weights (Step):", trained_weights)
    
```

```python

def Question_5():
    def sigmoid(x):
        return 1 / (1 + np.exp(-x))

    # Perceptron training function
    def train_perceptron(X, y, weights, alpha, epochs):
        for epoch in range(epochs):
            for i in range(len(X)):
                # Forward pass
                z = np.dot(X[i], weights)
                prediction = sigmoid(z)
                # Backpropagation
                error = y[i] - prediction
                weights += alpha * error * prediction * (1 - prediction) * X[i]
        return weights

    # Classify status 
    def classify_perceptron(data):
        # Load customer data
        purchase_data = data

        X = purchase_data.iloc[:, 1:-1].values
        y = purchase_data.iloc[:, -1].values

        X = (X - X.mean(axis=0)) / X.std(axis=0)

        X = np.c_[np.ones(X.shape[0]), X]

        weights = np.random.randn(X.shape[1])

        alpha = 0.1  # Learning rate
        epochs = 1000  # Number of training epochs

        weights = train_perceptron(X, y, weights, alpha, epochs)

        predictions = (sigmoid(np.dot(X, weights)) >= 0.5).astype(int)
        return predictions

    # Classify status using matrix pseudo-inverse
    def classify_pseudo_inverse(data):
        purchase_data = data
        X = purchase_data.iloc[:, 1:-1].values
        y = purchase_data.iloc[:, -1].values
        print(y)
        X = (X - X.mean(axis=0)) / X.std(axis=0)

        X = np.c_[np.ones(X.shape[0]), X]

        pseudo_inv = np.linalg.pinv(X)

        weights = np.dot(pseudo_inv, y)

        predictions = (sigmoid(np.dot(X, weights)) >= 0.5).astype(int)
        return predictions

    # Load customer data
    purchase_data = pd.read_excel('lab_3.xlsx', sheet_name = "Purchase data") # reading the excel fil e 
    customer_data = purchase_data[['Customer', 'Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)', 'Payment (Rs)', 'Status']]

    perceptron_predictions = classify_perceptron(customer_data)

    pseudo_inverse_predictions = classify_pseudo_inverse(customer_data)

    y_true = pd.read_excel('lab_3.xlsx', sheet_name = "Purchase data")["Status"].values
    print(y_true)

    perceptron_accuracy = accuracy_score(y_true, perceptron_predictions)
    pseudo_inverse_accuracy = accuracy_score(y_true, pseudo_inverse_predictions)

    print("Perceptron Accuracy:", perceptron_accuracy)
    print("Pseudo-Inverse Accuracy:", pseudo_inverse_accuracy)
```

```python
# Question 7:

# sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Forward pass function
def forward_pass(A, B, weights):
    # First layer
    v11 = sigmoid(np.dot([A, B], weights['W1'][0]))
    v12 = sigmoid(np.dot([A, B], weights['W1'][1]))
    v21 = sigmoid(np.dot([A, B], weights['W1'][2]))  
    v22 = sigmoid(np.dot([A, B], weights['W1'][3]))  
    
    # Second layer
    h1 = sigmoid(np.dot([v11, v21], weights['W2'][0]))  
    h2 = sigmoid(np.dot([v12, v22], weights['W2'][1]))  
    
    # Third layer
    O = sigmoid(np.dot([h1, h2], weights['W3']))
    
    return O

# Backpropagation function
def backpropagation(A, B, expected_output, weights, learning_rate):
    # Forward pass
    v11 = sigmoid(np.dot([A, B], weights['W1'][0]))
    v12 = sigmoid(np.dot([A, B], weights['W1'][1]))
    v21 = sigmoid(np.dot([A, B], weights['W1'][2]))  
    v22 = sigmoid(np.dot([A, B], weights['W1'][3]))  
    h1 = sigmoid(np.dot([v11, v21], weights['W2'][0]))
    h2 = sigmoid(np.dot([v12, v22], weights['W2'][1]))
    O = sigmoid(np.dot([h1, h2], weights['W3']))
    
    # Output layer error
    output_error = expected_output - O
    output_delta = output_error * sigmoid_derivative(O)
    
    # Hidden layer error
    hidden_error = weights['W3'] * output_delta
    hidden_delta1 = hidden_error[0] * sigmoid_derivative(h1)
    hidden_delta2 = hidden_error[1] * sigmoid_derivative(h2)
    
    # Update weights
    weights['W3'] += learning_rate * output_delta * np.array([h1, h2])
    weights['W2'][0] += learning_rate * hidden_delta1 * np.array([v11, v21])
    weights['W2'][1] += learning_rate * hidden_delta2 * np.array([v12, v22])
    weights['W1'][0] += learning_rate * hidden_delta1 * np.array([A, B])
    weights['W1'][1] += learning_rate * hidden_delta2 * np.array([A, B])
    weights['W1'][2] += learning_rate * hidden_delta1 * np.array([A, B])  
    weights['W1'][3] += learning_rate * hidden_delta2 * np.array([A, B])  
    
    return weights

# Training the neural networks 
def train_neural_network(inputs, expected_outputs, weights, learning_rate, epochs):
    errors = []  # store the  sum squared error at each epoch
    for epoch in range(epochs):
        epoch_error = 0
        for i in range(len(inputs)):
            # Forward pass
            output = forward_pass(inputs[i][0], inputs[i][1], weights)
            # Compute error
            error = expected_outputs[i] - output
            epoch_error += error ** 2
            # Backpropagation
            weights = backpropagation(inputs[i][0], inputs[i][1], expected_outputs[i], weights, learning_rate)
        errors.append(epoch_error)
    return weights, errors

# initial weights
initial_weights = {
    'W1': [np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5])],
    'W2': [np.array([0.5, 0.5]), np.array([10, 0.5])],
    'W3': np.array([0.5, 0.5])
}

# training data (AND gate)
inputs_AND = [(0, 0), (0, 1), (1, 0), (1, 1)]
expected_outputs_ANd = [0, 0, 0, 1]

# Training  the neural network
trained_weights, errors = train_neural_network(inputs_AND, expected_outputs_ANd, initial_weights, learning_rate=0.01, epochs=10000)

# Plot epoch vs. sum squared error
plt.plot(range(1, len(errors) + 1), errors)
plt.xlabel('Epochs')
plt.ylabel('Sum Squared Error')
plt.title('Epochs vs. Sum Squared Error')
plt.show()
```

```python


# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Forward pass function
def forward_pass(A, B, weights):
    # First layer
    v11 = sigmoid(np.dot([A, B], weights['W1'][0]))
    v12 = sigmoid(np.dot([A, B], weights['W1'][1]))
    v21 = sigmoid(np.dot([A, B], weights['W1'][2]))  
    v22 = sigmoid(np.dot([A, B], weights['W1'][3]))  
    
    # Second layer
    h1 = sigmoid(np.dot([v11, v21], weights['W2'][0]))  
    h2 = sigmoid(np.dot([v12, v22], weights['W2'][1]))
    h3 = sigmoid(np.dot([v11, v21], weights['W2'][2]))  
    h4 = sigmoid(np.dot([v12, v22], weights['W2'][3]))
    
    # Third layer
    O1 = sigmoid(np.dot([h1, h3], weights['W3'][0]))
    O2 = sigmoid(np.dot([h2, h4], weights['W3'][1]))
    
    return O1, O2

# Backpropagation function
def backpropagation(A, B, expected_output, weights, learning_rate):
    # Forward pass
    v11 = sigmoid(np.dot([A, B], weights['W1'][0]))
    v12 = sigmoid(np.dot([A, B], weights['W1'][1]))
    v21 = sigmoid(np.dot([A, B], weights['W1'][2]))  
    v22 = sigmoid(np.dot([A, B], weights['W1'][3]))  
    h1 = sigmoid(np.dot([v11, v21], weights['W2'][0]))
    h2 = sigmoid(np.dot([v12, v22], weights['W2'][1]))
    h3 = sigmoid(np.dot([v11, v21], weights['W2'][2]))  
    h4 = sigmoid(np.dot([v12, v22], weights['W2'][3]))
    O1 = sigmoid(np.dot([h1, h3], weights['W3'][0]))
    O2 = sigmoid(np.dot([h2, h4], weights['W3'][1]))
    
    # Output layer error
    output_error1 = expected_output[0] - O1
    output_delta1 = output_error1 * sigmoid_derivative(O1)
    
    output_error2 = expected_output[1] - O2
    output_delta2 = output_error2 * sigmoid_derivative(O2)
    
    # Hidden layer error
    hidden_error1 = weights['W3'][0] * output_delta1
    hidden_delta11 = hidden_error1[0] * sigmoid_derivative(h1)
    hidden_delta31 = hidden_error1[1] * sigmoid_derivative(h3)
    
    hidden_error2 = weights['W3'][1] * output_delta2
    hidden_delta12 = hidden_error2[0] * sigmoid_derivative(h2)
    hidden_delta32 = hidden_error2[1] * sigmoid_derivative(h4)
    
    # Update weights
    weights['W3'][0] += learning_rate * output_delta1 * np.array([h1, h3])
    weights['W3'][1] += learning_rate * output_delta2 * np.array([h2, h4])
    
    weights['W2'][0] += learning_rate * hidden_delta11 * np.array([v11, v21])
    weights['W2'][1] += learning_rate * hidden_delta12 * np.array([v12, v22])
    weights['W2'][2] += learning_rate * hidden_delta31 * np.array([v11, v21])
    weights['W2'][3] += learning_rate * hidden_delta32 * np.array([v12, v22])
    
    weights['W1'][0] += learning_rate * hidden_delta11 * np.array([A, B])
    weights['W1'][1] += learning_rate * hidden_delta12 * np.array([A, B])
    weights['W1'][2] += learning_rate * hidden_delta31 * np.array([A, B])
    weights['W1'][3] += learning_rate * hidden_delta32 * np.array([A, B])
    
    return weights

# Training the neural networks 
def train_neural_network(inputs, expected_outputs, weights, learning_rate, epochs):
    errors = []  # store the sum squared error at each epoch
    for epoch in range(epochs):
        epoch_error = 0
        for i in range(len(inputs)):
            # Forward pass
            output1, output2 = forward_pass(inputs[i][0], inputs[i][1], weights)
            # Compute error
            error1 = expected_outputs[i][0] - output1
            error2 = expected_outputs[i][1] - output2
            epoch_error += (error1 ** 2 + error2 ** 2)
            # Backpropagation
            weights = backpropagation(inputs[i][0], inputs[i][1], (expected_outputs[i][0], expected_outputs[i][1]), weights, learning_rate)
        errors.append(epoch_error)
    return weights, errors

# Initial weights
initial_weights = {
    'W1': [np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5])],
    'W2': [np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5]), np.array([0.5, 0.5])],
    'W3': [np.array([0.5, 0.5]), np.array([0.5, 0.5])]
}

# Training data (AND gate)
inputs_AND = [(0, 0), (0, 1), (1, 0), (1, 1)]
expected_outputs_AND = [(0, 0), (0, 0), (0, 0), (1, 1)]

# Training the neural network
trained_weights, errors = train_neural_network(inputs_AND, expected_outputs_AND, initial_weights, learning_rate=0.01, epochs=10000)

# Plot epoch vs. sum squared error
plt.plot(range(1, len(errors) + 1), errors)
plt.xlabel('Epochs')
plt.ylabel('Sum Squared Error')
plt.title('Epochs vs. Sum Squared Error')
plt.show()
```

```python
# AND gate data
X_and = [[0, 0], [0, 1], [1, 0], [1, 1]]
y_and = [0, 0, 0, 1]

# XOR gate data
X_xor = [[0, 0], [0, 1], [1, 0], [1, 1]]
y_xor = [0, 1, 1, 0]
max_iter_values = range(100, 2001, 100)  # Max_iter values from 100 to 2000 in steps of 100

# Lists to store errors for AND gate and XOR gate
errors_and = []
errors_xor = []

# Training MLP classifiers for different max_iter values
for max_iter in max_iter_values:
    # MLPClassifier for AND gate
    mlp_and = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', max_iter=max_iter)
    mlp_and.fit(X_and, y_and)
    errors_and.append(1 - mlp_and.score(X_and, y_and))  # Append error for AND gate
    
    # MLPClassifier for XOR gate
    mlp_xor = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', max_iter=max_iter)
    mlp_xor.fit(X_xor, y_xor)
    errors_xor.append(1 - mlp_xor.score(X_xor, y_xor))  # Append error for XOR gate

# Plotting max_iter vs. error
plt.plot(max_iter_values, errors_and, label='AND Gate')
plt.plot(max_iter_values, errors_xor, label='XOR Gate')
plt.xlabel('Max Iterations')
plt.ylabel('Error')
plt.title('Max Iterations vs. Error')
plt.legend()
plt.show()

# predictions for AND gate
print("AND Gate Predictions:")
for i in range(len(X_and)):
    prediction = mlp_and.predict([X_and[i]])
    print(f"Input: {X_and[i]}, Predicted Output: {prediction}")

# predictions for XOR gate
print("\nXOR Gate Predictions:")
for i in range(len(X_xor)):
    prediction = mlp_xor.predict([X_xor[i]])
    print(f"Input: {X_xor[i]}, Predicted Output: {prediction}")
```
